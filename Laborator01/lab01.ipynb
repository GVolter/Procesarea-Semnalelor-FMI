EX1. Fie semnalele continue x(t)=cos (520 pi t+pi / 3), y(t)=cos (280 pi t-pi / 3) și z(t)=cos (120 pi t+pi / 3)

(a) În Python, simulați axa reală de timp printr-un șir de numere suficient de apropiate, spre exemplu [0:0.0005:0.03].

```python
t = np.arange(0, 0.03, 0.0005) 
```

(b) Construiți semnalele x(t), y(t) și z(t) și afișați-le grafic, în cate un subplot.

```python
x = lambda t: np.cos(520 * np.pi * t + np.pi / 3)
y = lambda t: np.cos(280 * np.pi * t - np.pi / 3)
z = lambda t: np.cos(120 * np.pi * t + np.pi / 3)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 8), constrained_layout=True)

axs[0].plot(t, x(t))
axs[0].set_title('x(t)')

axs[1].plot(t, y(t), color = "r")
axs[1].set_title('y(t)')

axs[2].plot(t, z(t), color = "purple")
axs[2].set_title('z(t)')

plt.show()
```

(c) Eșantionați semnalele cu o frecvență de 200 Hz pentru a obține x[n], y[n] și z[n] și afișați-le grafic, în câte un subplot.

```python
freq = 200
n = np.linspace(0, 0.03, freq)

fig, axs = plt.subplots(3, 1, figsize=(10, 8), constrained_layout=True)

axs[0].stem(n, x(n))
axs[0].set_title('x[n]')

axs[1].stem(n, y(n), 'red')
axs[1].set_title('y[n]')

axs[2].stem(n, z(n), 'purple')
axs[2].set_title('z[n]')

plt.show()
```

EX2. Generați următoarele semnale și afișați-le grafic, fiecare într-un plot:

(a) Un semnal sinusoidal de frecvență 400Hz, care să conțină 1600 de esantioane.

```python
def sin_wave(t, freq):
    return np.sin(2 * np.pi * freq *t)
```

```python
freq = 400
t = np.linspace(0, 2/freq , 1600)

plt.plot(t, sin_wave(t, freq))
plt.suptitle("Semnal sinusoidal 400Hz")
plt.show()
```

(b) Un semnal sinusoidal de frecvență 800Hz, care să dureze 3 secunde.

```python
freq = 800
duration = 3
t = np.linspace(0, duration, 3 * freq)

plt.plot(t, sin_wave(t, freq))
plt.title("Semnal sinusoidal 800Hz")
plt.show()
```

(c) Un semnal de tip sawtooth de frecvență 240Hz (puteți folosi funcțiile numpy.floor sau numpy.mod).

```python
def sawtooth(t, freq):
    return 2 * (freq * t - np.floor(240 * t)) - 1
```

```python
freq = 240
t = np.linspace(0, 4/freq, 480)  # Reusing the defined duration and sample rate

plt.figure()
plt.plot(t, sawtooth(t, freq))
plt.title('Semnal Sawtooth')
plt.show()
```

(d) Un semnal de tip square de frecvență 300Hz (puteți folosi funcția numpy.sign).

```python
def square(t, freq):
    return np.sign(np.sin(2 * np.pi * freq *t))

freq = 300
t = np.linspace(0, 2/freq, freq)

plt.plot(t, square(t, freq))
plt.title('Semnal Square')
plt.show()
```

(f) Un semnal 2D la alegerea voastră. Creați un numpy.array de dimensiune 128x128 și inițializați-l folosind o procedură creată de voi. Utilizați, spre exemplu, funcțiile numpy.zeros () și numpy.ones ().

```python
m = np.zeros((128, 128))

num_circles = 8

step = 64 / num_circles

for radius in np.linspace(step, 64, num_circles):
    for i in range(128):
        for j in range(128):
            distance = np.sqrt((i - 64) ** 2 + (j - 64) ** 2)
            if int(distance / step) == int(radius / step):
                m[i, j] = (int(radius / step)) % 2

plt.figure()
plt.imshow(m)
plt.title('Semnal 2D')
plt.show()
```
